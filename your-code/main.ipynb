# StackAPI

#### Import the necessary libraries here:

```python
import pandas as pd
import numpy as np
from stackapi import StackAPI


SITE = StackAPI('stackoverflow')

comments = SITE.fetch('comments')

comments_df = pd.DataFrame(comments['items'])
comments_df.head()
```

#### Question 1: Find the questions and answers of last month.

```python
questions = SITE.fetch('questions', fromdate = 1570532113, todate = 1573214113)
questions_df = pd.DataFrame(questions['items'])
questions_df.head()
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

we search with the tag with the word python

```python
questions = SITE.fetch('questions', todate = 1573214113,tagged = 'python')
questions_df = pd.DataFrame(questions['items'])
```

we localiza the questions tagged by python with a score >= 5 and sorted descending

```python
questions_df.loc[(questions_df.score >= 5)].sort_values('score', ascending = False).head()
```
